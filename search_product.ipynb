```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
flipkart_scraper = AutoScraper()
flipkart_scraper.load('flipkart.json')
```

```python
search = "curl hair care" 
flipkart_url="https://www.flipkart.com/search?q={}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=popularity".format(search)

data = flipkart_scraper.get_result_similar(flipkart_url, group_by_alias=True)
search_data = tuple(zip(data['Title'],data['ImageUrl'],data['Price'],data['Reviews']))
```

```python
df = pd.DataFrame(columns=['Query','Title','Price','Reviews','ImageUrl'])
for i in range(len(search_data)):
    df.loc[len(df)] = [search,search_data[i][0],search_data[i][2],search_data[i][3],search_data[i][1]]
```
